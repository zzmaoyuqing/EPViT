```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar
```

## input

```python
xlsx_file = pd.ExcelFile('result.xlsx')
```

```python
df_5093 = pd.read_excel(xlsx_file,'PPI5093',index_col=0)
df_5093 = df_5093.iloc[:,0:4]
df_5093
```

```python
df_3672 = pd.read_excel(xlsx_file,'PPI3672',index_col=0)
df_3672 = df_3672.iloc[:,0:4]
df_3672
```

```python
df_2708 = pd.read_excel(xlsx_file,'PPI2708',index_col=0)
df_2708 = df_2708.iloc[:,0:4]
df_2708
```

## PPI5093

```python
# original: '#C52A20','#A2C8DD','#7F7DB5','#A1D0C7'-->after:'#C52A20','#96CCCB','#7F7DB5''#F3D266'
bar1 = (
    Bar()
    .add_xaxis(df_5093.columns.to_list())
    .add_yaxis('EPViT',df_5093.iloc[0,:].to_list(),color='#C52A20')
    .add_yaxis('MBIEP',df_5093.iloc[1,:].to_list(),color='#96CCCB')
    .add_yaxis('DeepEP',df_5093.iloc[2,:].to_list(),color='#7F7DB5')
    .add_yaxis('[16]',df_5093.iloc[3,:].to_list(), color='#F3D266')

    .set_global_opts(xaxis_opts=opts.AxisOpts(name='metrics'))
    .set_global_opts(yaxis_opts=opts.AxisOpts(name='values'))
    .set_global_opts(legend_opts=opts.LegendOpts(pos_left='right',
                                                 orient='horizontal'))
    .set_global_opts(title_opts=opts.TitleOpts(pos_left='center',
                                               pos_top=20))
#     .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_series_opts(label_opts=opts.LabelOpts(position='top',
                                              font_size=10))
    #  ToolboxOpts:Toolbox Configurations
    .set_global_opts(toolbox_opts=opts.ToolboxOpts(is_show = True,
                                                     pos_top = 'bottom',
                                                     pos_left = 'right',
                     # 5 represents 5 times the resolution, the default is 1, the bigger the number, the clearer the saved picture is. 
                     feature={'saveAsImage':{'pixelRatio':5}}))

)
```

```python
bar1.render_notebook()
```
