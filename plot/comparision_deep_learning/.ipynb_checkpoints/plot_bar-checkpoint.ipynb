```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar
```

## input

```python
xlsx_file = pd.ExcelFile('result.xlsx')
```

```python
df_5093 = pd.read_excel(xlsx_file,'PPI5093',index_col=0)
df_5093 = df_5093.iloc[:,0:4]
df_5093
```

```python
df_3672 = pd.read_excel(xlsx_file,'PPI3672',index_col=0)
df_3672 = df_3672.iloc[:,0:4]
df_3672
```

```python
df_2708 = pd.read_excel(xlsx_file,'PPI2708',index_col=0)
df_2708 = df_2708.iloc[:,0:4]
df_2708
```

# Plot bar and Click the 'Save Image' button
## There are two types of cases: 
### 1.adjusting the starting scale value 
### 2.not adjusting the starting scale value.            DELETE min_=0.2, interval=0.2,

## 1.adjusting the starting scale value 

## PPI5093

```python
bar1_a = (
    Bar()
    .add_xaxis(df_5093.columns.to_list())
    .add_yaxis('EPViT', df_5093.iloc[0,:].apply(lambda x: f'{x:.3f}').to_list(), color='#C52A20')
    .add_yaxis('MBIEP', df_5093.iloc[1,:].apply(lambda x: f'{x:.3f}').to_list(), color='#96CCCB')
    .add_yaxis('DeepEP', df_5093.iloc[2,:].apply(lambda x: f'{x:.3f}').to_list(), color='#7F7DB5')
    .add_yaxis('Zeng et al., 2021', df_5093.iloc[3,:].apply(lambda x: f'{x:.3f}').to_list(), color='#F3D266')
    
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name='metrics',
            name_textstyle_opts=opts.TextStyleOpts(font_size=20),  # Set x-axis name font size
            axislabel_opts=opts.LabelOpts(font_size=25)  # Set x-axis label font size
        ),
        yaxis_opts=opts.AxisOpts(
            name='values',
            name_textstyle_opts=opts.TextStyleOpts(font_size=20),  # Set y-axis name font size
            min_=0.2,                 # Set y-axis minimum value to 0.2
            interval=0.2,             # Set y-axis tick interval to 0.2
            axislabel_opts=opts.LabelOpts(font_size=16)  # Set y-axis label font size
        ),
        legend_opts=opts.LegendOpts(
            pos_left='right', 
            orient='horizontal',
            textstyle_opts=opts.TextStyleOpts(font_size=25)  # Set legend font size
        ),
        title_opts=opts.TitleOpts(pos_left='center', pos_top=20),
        toolbox_opts=opts.ToolboxOpts(
            is_show=True,
            pos_top='bottom',
            pos_left='right',
            feature={'saveAsImage': {'pixelRatio': 5}}
        )
    )
        
    .set_series_opts(
        label_opts=opts.LabelOpts(
            position='top', 
            font_size=15,
            color='black'
        )
    )
    
)
```

```python
bar1_a.render_notebook()
```
